```python
#importing essential libraries and plotting Signal
import numpy as np  
import scipy as sp  
from scipy.io.wavfile import read  
from scipy.io.wavfile import write 
from scipy.fftpack import fft
from scipy import signal  
import matplotlib.pyplot as plt  
get_ipython().magic('matplotlib inline')  

import librosa
(array, Frequency) = librosa.load(r"C:\Users\Anusha\Downloads\Breathing-sound.wav",sr=None) 

len(array)  

plt.plot(array)   
plt.title('Original Signal Spectrum')  
plt.xlabel('Frequency(Hz)')  
plt.ylabel('Amplitude')  

```

```python
#Determining FFT of signal and plotting
FourierTransformation = fft(array)  

scale = sp.linspace(0, Frequency, len(array))  

plt.stem(scale, np.abs(FourierTransformation), 'r')    
plt.title('Signal spectrum after FFT')  
plt.xlabel('Frequency(Hz)')  
plt.ylabel('Amplitude')  


```

```python
#Applying FIR Filter with hamming window operation for better Stopband Attenuation(Still Incomplete)
from scipy import signal
import numpy as np
b = signal.firwin(80, 0.3283728, window='hamming',pass_zero='lowpass')#Cut off calculated from frequency response of Original Signal
y = np.array([np.convolve(array, b, mode='valid')])
plt.plot(y)   
plt.title('Filtered Signal Spectrum')  
plt.xlabel('Frequency(Hz)')  
plt.ylabel('Amplitude')  

```

```python
#Gaining Frequency band information of Filtered Signal
y_fft=fft(y)
scale = sp.linspace(0, Frequency, len(y))  

plt.stem(scale, np.abs(y_fft), 'r')    
plt.title('NewSignal spectrum after FFT')  
plt.xlabel('Frequency(Hz)')  
plt.ylabel('Amplitude')  
```
